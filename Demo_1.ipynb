# nlargest() and nsmallest()

```python
df.nlargest(10,columns=['NA_Sales']).set_index('Name').loc[:,'NA_Sales'].plot.bar()
plt.title('Japan',fontsize=20)
plt.show()plt.title('North America',fontsize=20)
plt.show()
```

```python
df.nlargest(10,columns=['EU_Sales']).set_index('Name').loc[:,'EU_Sales'].plot.bar()
plt.title('Europe',fontsize=20)
plt.show()
```

```python
df.nlargest(10,columns=['JP_Sales']).set_index('Name').loc[:,'JP_Sales'].plot.bar()
plt.title('Japan',fontsize=20)
plt.show()
```

```python
df.nlargest(10,columns=['Other_Sales']).set_index('Name').loc[:,'Other_Sales'].plot.bar()
plt.title('Other',fontsize=20)
plt.show()
```

```python
df.nlargest(10,columns=['Global_Sales']).set_index('Name').loc[:,'Global_Sales'].plot.bar()
plt.title('Global',fontsize=20)
plt.show()
```

```python
fig, ax =plt.subplots(2,2, figsize = (30,50))

df.nlargest(10,columns=['JP_Sales']).set_index('Name').loc[:,'JP_Sales'].plot.bar(ax=ax[0][0])
ax[0][0].set_title('Japan', fontsize = 20)
df.nlargest(10,columns=['EU_Sales']).set_index('Name').loc[:,'EU_Sales'].plot.bar(ax=ax[0][1])
ax[0][1].set_title('Europe', fontsize = 20)
df.nlargest(10,columns=['NA_Sales']).set_index('Name').loc[:,'NA_Sales'].plot.bar(ax=ax[1][0])
ax[1][0].set_title('North America', fontsize = 20)
df.nlargest(10,columns=['Other_Sales']).set_index('Name').loc[:,'Other_Sales'].plot.bar(ax=ax[1][1])
ax[1][1].set_title('Other', fontsize = 20)

# fig.subtitle('Most Game Sold Continent Wise', fontsize = 20)
fig.tight_layout()
plt.show()
```
